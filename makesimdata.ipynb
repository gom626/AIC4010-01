```
import matplotlib.pyplot as plt
import cv2
import re
import numpy as np
```

```
fig, ax = plt.subplots(figsize=(20,15))
img = cv2.imread('predictions.jpg', cv2.IMREAD_COLOR)
b, g, r = cv2.split(img)
img = cv2.merge([r,g,b])
ax.imshow(img, aspect='auto')
```

```
fp = open('/content/darknet/result.txt', 'r')

data = []
shoes = {}
for line in fp.readlines():
  if "Enter Image Path" in line:
    if shoes != {}:
      data.append(shoes)
      shoes = {}
    s_idx = 18
    e_idx = line.find(': Predicted')
    img_path = line[s_idx:e_idx]
    img_name = img_path.split('/')[-1][:-4]
    shoes['img_path'] = img_path
    shoes['img_name'] = img_name
    shoes['box'] = []
  elif "shoes:" in line:
    percent = line[7:9]
    prob = int(percent)
    if prob < 50:
      continue
    box = {}
    idx1 = line.find('left_x:')
    idx2 = line.find('top_y:')
    idx3 = line.find('width:')
    idx4 = line.find('height:')
    idx5 = line.find(')')

    box['left_x'] = int(line[idx1+7:idx2].strip())
    box['top_y'] = int(line[idx2+6:idx3].strip())
    box['width'] = int(line[idx3+6:idx4].strip())
    box['height'] = int(line[idx4+7:idx5].strip())
    shoes['box'].append(box)
```

```
data
!mkdir output
```

```
for d in data:
  img = cv2.imread(d['img_path'])
  img_name = d['img_name']
  for idx, box in enumerate(d['box']):
    x1 = box['left_x']
    x2 = x1 + box['width']
    y1 = box['top_y']
    y2 = y1 + box['height']

    shoes = img[y1:y2, x1:x2, :]
    height = shoes.shape[0]
    width = shoes.shape[1]
    dif = np.abs(height - width) // 2
    if height > width:
      shoes_pad = np.pad(shoes, pad_width = ((0,0),(dif, dif), (0,0)), constant_values=255)
    elif height < width:
      shoes_pad = np.pad(shoes, ((dif, dif),(0,0), (0,0)), constant_values=255)
    
    cv2.imwrite('/content/darknet/output/'+img_name+'_'+str(idx)+'.jpg', shoes_pad)
```
